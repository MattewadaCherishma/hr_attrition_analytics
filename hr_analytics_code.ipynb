```python
#coverting attrition to numbers(for correlation)
df['AttritionFlag']=(df['Attrition']=='Yes').astype(int)
```

```python
#correlation analysis
numeric_cols = ['Age','MonthlyIncome','JobSatisfaction','WorkLifeBalance',
                 'YearsAtCompany','DistanceFromHome','AttritionFlag']
df[numeric_cols].corr()['AttritionFlag'].sort_values()
```

```python
import matplotlib.pyplot as plt
df.groupby('OverTime')['AttritionFlag'].mean().mul(100).plot(kind='bar', color=['#2563EB','#DC2626'])
plt.ylabel('Attrition Rate (%)')
plt.title('Attrition Rate by OverTime Status')
plt.show()
```

```python
df.groupby('Department')['AttritionFlag'].mean().mul(100).plot(kind='bar', color=['#2563EB','#DC2626'])
plt.ylabel('Attrition Rate (%)')
plt.title('Attrition Rate by OverTime Status')
plt.show()
```
